## Groupby Aggregation

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
df.columns
df['Diff_POPESTIMATE_2011_2010'] = df['POPESTIMATE2011'] -df['POPESTIMATE2010']
```

```python
df.head()
```

```python
df['SUMLEV'].unique()
```

```python
df.groupby('SUMLEV').agg({'Diff_POPESTIMATE_2011_2010': np.average})
```

```python
df.groupby('SUMLEV').agg({'Diff_POPESTIMATE_2011_2010': (np.nanmean,np.nanstd),'POPESTIMATE2010':np.average})
```

## Transformation

```python
cols = ['SUMLEV','Diff_POPESTIMATE_2011_2010']

transform_df = df[cols].groupby('SUMLEV').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'Diff_POPESTIMATE_2011_2010':'Mean_Diff_POPESTIMATE_2011_2010'},axis = 'columns',inplace = True)
transform_df.head()
```

```python
df = df.merge(transform_df,left_index = True,right_index = True)
df.head()
```

## Filter

```python
df.groupby('SUMLEV').filter(lambda x: np.nanmean(x['Diff_POPESTIMATE_2011_2010'])> 780)
```

```python
a = lambda x: np.nanmean(x)
a(5)
```

## Pivot table

```python
df = pd.read_csv('datasets/cwurData.csv')
df.head()
```

```python
df['national_rank'].unique()
```

Rank is between 1 to 100 then First Class institution.

Rank is between 100 to 200 then Second Class institution.

Rank is between 200 to 300 then Other institution.


```python
def create_category(rank):
    if (rank>=1) & (rank<=100):
        return 'First Tier Top Univeristy'
    elif (rank>=101) & (rank<=200):
        return 'Second Tier Top University'
    else:
        return 'Other Top University'
```

```python
df['Rank_Level'] = df['national_rank'].apply(create_category)
```

```python
df.head()
```

```python
df.pivot_table(values = 'score',index = 'country',columns = 'Rank_Level',aggfunc = [np.mean]).head()
```

```python
df.pivot_table(values = 'score',index = 'country',columns = 'Rank_Level',aggfunc = [np.mean,np.max]).head()
```

```python
new_df =df.pivot_table(values = 'score',index = 'country',columns = 'Rank_Level',aggfunc = [np.mean,np.max],margins = True)
```

```python
new_df['mean']['First Tier Top Univeristy']
```

```python
new_df['mean']['First Tier Top Univeristy'].idxmax()
```

## Stacking

```python
new_df = new_df.stack()
new_df
```

## Unstacking

```python
new_df = new_df.unstack()
new_df
```

```python
new_df.unstack().unstack().head()
```
